# Concatenação de DataFrames

```python
# Criando 2 DataFrames para concatenar
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
df1 = pd.DataFrame(data) 
df1
```

```python
# Configura os índices do DataFrame
indice = ['Linha ' + str(i) for i in range(len(data))]
coluna = ['Coluna ' + str(i) for i in range(len(data))]
df1 = pd.DataFrame(data=data, index=indice, columns=coluna)
df1
```

```python
dicio = {
    'Coluna 0' : {'Linha 0' : 1, 'Linha 1': 4, 'Linha 2' : 7},
    'Coluna 1' : {'Linha 0' : 2, 'Linha 1': 5, 'Linha 2' : 8},
    'Coluna 2' : {'Linha 0' : 3, 'Linha 1': 6, 'Linha 2' : 9}
}
df2 = pd.DataFrame(dicio)
df2
```

```python
# É possível alterar os valores
df1[df1 > 0] = 'A'
df2[df2 > 0] = 'B'
df1
```

```python
# concat() empilha os DataFrames, mas também é possível pô-los alinhados
df3 = pd.concat([df1, df2])
df3
```

```python
df3 = pd.concat([df1, df2], axis=1)
df3
```

# Removendo Dados Nulos

```python
import pandas as pd
dados = pd.read_csv('aluguel.csv', sep=';')
dados
```

```python
residenciais = ['Quitinete', 'Casa', 'Apartamento', 'Casa de Condomínio', 'Casa de Vila']
selecao = dados.Tipo.isin(residenciais)
dados_residenciais = dados[selecao]
dados_residenciais
```

```python
# notnull() faz um DataFrame de True or False, se uma célula é vazia o valor será False
# isnull() é o inverso
dados_residenciais.notnull()
```

```python
# Lista um DataFrame com valor de 'Valor' nulo
dados_residenciais[dados_residenciais['Valor'].isnull()]
```

```python
# Removendo valores nulos
nao_nulos = dados_residenciais.dropna(subset=['Valor'])
nao_nulos
```

```python
selecao = (dados_residenciais['Tipo'] == 'Casa') & (dados_residenciais['IPTU'].notnull())
dados = dados_residenciais[selecao] 
# dados_residenciais[~selecao] inverte o resultado → !selecao
dados
```

```python
# Preencher os nulos com outros valores
# Preencher com zeros (ou outro valor especificado)
fill = dados_residenciais.IPTU.fillna(0)
fill
```

```python
# Preencher com o último valor válido acima
dados_residenciais.IPTU.fillna(method='ffill')
```

```python
# Preencher com último valor válido abaixo
dados_residenciais.IPTU.fillna(method='bfill')
```

```python
# É possível limitar quantas vezes o valor será copiado
a = dados_residenciais.IPTU.fillna(method='ffill', limit=1)
# ou atribuir a média dos valores
b = dados_residenciais.IPTU.fillna(dados_residenciais.IPTU.mean())
print(a, b)
```

# Incluindo novas Variáveis

```python
dados = pd.read_csv('aluguel_residencial.csv', sep=';')
dados.head()
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
dados['Tipo Imovel'] = dados['Tipo'].apply(lambda x : 'Casa' if x in casa else 'Apartamento')
dados
```

```python
dados['Valor Bruto'] = (dados['Valor'] + dados['Condominio'] + dados.Condominio).round(2)
dados
```

# Excluindo Variáveis

```python
# Há algumas formas de exclusão: del, .pop() e .drop()
# dados_aux = dados
dados_aux = pd.DataFrame(dados[['Area', 'Tipo', 'Valor Bruto', 'Tipo Imovel']])
dados_aux
```

```python
del(dados_aux['Valor Bruto'])
dados_aux
```

```python
dados_aux.pop('Tipo Imovel')
dados_aux.head()
```

```python
dados.drop(['Tipo Imovel', 'Valor Bruto'], axis=1, inplace=True)
dados.head()
```

```python
df = pd.read_csv("bases/MODULO7_PROJETO_FINAL_BASE_SUPERMERCADO.csv", delimiter=',')
df.head()
```

Os campos do nosso dataframe são:

**Title:** Nome do produto.


**Marca:** A marca do produto.


**Preco_Normal:** O preço em que o produto costuma ser vendido quando não há desconto.


**Preco_Desconto:** O preço vendido após o desconto ser aplicado.


**Preco_Anterior:** Preço em que era comercializado o produto antes do desconto aplicado.


**Desconto:** Total de desconto aplicado.






As colunas que aparecem com valores 0 são para os produtos onde não tivemos descontos aplicados.


As categorias estão em espanhol!

# 1 - Traga a média e a mediana dos preços - coluna Preco_Normal - por categoria de produto.
# Identifique as categorias que parecem ter um valor de média abaixo ou acima da mediana.

```python
#Seu código aqui para a média
media = df.groupby("Categoria")["Preco_Normal"].mean().round(2)
media
```

```python
#Seu código aqui para a mediana
mediana = df.groupby("Categoria")["Preco_Normal"].median().round(2)
mediana
```

Digite aqui as categorias:

```python
acima = " ".join(media[media > mediana].index)
abaixo = " ".join(media[media < mediana].index)
print(f"Categorias com médias maiores que a mediana: {acima}.")
print(f"Categorias com médias menores que a mediana: {abaixo}.")
```

# 2 - Traga o desvio padrão por categoria de produto.
# Qual o comportamento da média e mediana nas categorias com maior desvio?

```python
#Seu código aqui
desvio = df.groupby("Categoria")["Preco_Normal"].std().sort_values(ascending=False)
desvio
```

```python
estat = pd.DataFrame(columns=["Categoria", "Média", "Mediana", "Desvio"])
estat["Categoria"] = desvio.index.sort_values()
estat["Média"] = estat["Categoria"].map(media)
estat["Mediana"] = estat["Categoria"].map(mediana)
estat["Desvio"] = estat["Categoria"].map(desvio)
estat
```

```python
estat.loc[:, "CV"] = (estat["Desvio"] / estat["Média"]).round(4)
estat
```

---

Este método foi feito apenas para exibir os gráficos no navegador, pois o VSCode não estava conseguindo exibir.

```python
import os
import webbrowser


def plot_to_html(fig, name: str):
    caminho_arquivo = os.path.abspath(name)
    fig.write_html(caminho_arquivo)
    
    comando_incognito = "firefox --private-window %s" 
    browser = webbrowser.get(comando_incognito)
    
    browser.open(caminho_arquivo)
```

# 3 - Plot um boxplot da distribuição do Preco_Normal para a categoria que você identificou que tem o maior desvio padrão. Como é a distribuição desses dados segundo o boxplot? Você identifica muitos outliers?

Dica: Para trazer apenas os dados da categoria que você deseja você pode usar o df.loc[df['Categoria'] == 'CATEGORIA ESCOLHIDA'

```python
#Seu código aqui
categorias = "belleza-y-cuidado-personal,instantaneos-y-sopas,lacteos"
cat_interesse = df.loc[df["Categoria"].isin(categorias.split(","))]

fig = px.box(
    cat_interesse, 
    x="Categoria", 
    y="Preco_Normal", 
    title="Categorias com alto Desvio Padrão | Presença de Outliers", 
    color="Categoria", 
    color_discrete_sequence=["#ff033e", "#9966cc", "#ffbf00"]
)
plot_to_html(fig, "box_preco.html")
```

# 4 - Plote um gráfico de barras onde temos a média de descontos por categoria.

```python
# Seu código aqui
desc_cat = df.groupby("Categoria")["Desconto"].mean().reset_index()
desc_cat = desc_cat.sort_values("Desconto", ascending=False)
fig = px.bar(
    desc_cat, 
    x="Categoria", 
    y="Desconto", 
    title="Média de Descontos por Categoria", 
    color="Categoria", 
)
plot_to_html(fig, "descontos.html")
```

# 5 - Plote um gráfico de mapa interativo agrupando os dados por categoria, marca e trazendo a média de desconto.

```python
# Seu código aqui
fig = px.density_heatmap(
    df, 
    x="Categoria", 
    y="Marca", 
    z="Desconto", 
    histfunc="avg", 
    color_continuous_scale=["#f4bbff", "#800020"], 
    title="Média de Desconto por Categoria e Marca"
)
plot_to_html(fig, "heatmap_desconto.html")
```
